## Folded section #1

```python
%matplotlib widget
import numpy as np
from bmcs_cross_section.api import ReinfLayer, BarLayer, MKappa, TShape
mc = MKappa(low_kappa=0, high_kappa=0.0001)
mc.cs_design.cross_section_shape_.trait_set(H=800, B=300)
mc.cs_design.concrete.compression = 'EC2 with plateau'
```

```python
mc.cs_design.cross_section_shape = 'custom'
thickness = 15 # mm
H = 125 # mm
B_lower = 71.5 # mm
B_upper = 250 + B_lower # mm
theta = np.pi / 2 * 1 / 2
```

```python
gamma = (np.pi - theta) / 2
B_lower_top = B_lower - thickness / np.tan(gamma)
```

```python
mc.cs_design.cross_section_shape_.cs_points = np.array([
    [B_lower_top, thickness],
    [0, thickness],
    [0, 0],
    [B_lower, 0],
    [B_lower + (H - thickness) / np.tan(theta), H - thickness ], 
    [B_lower + (H - thickness) / np.tan(theta) + B_upper, H - thickness ], 
    [B_lower + (H - thickness) / np.tan(theta) + B_upper, H ], 
    [B_lower_top + + (H - thickness) / np.tan(theta), H], 
])
```

```python
mc.cs_design.cross_section_shape_.interact()
```

```python
mc.cs_design.concrete.compression_.trait_set(
    f_c = 70.8, eps_cy=-0.0035, eps_cu=-0.0035, E_cc=25000
)
```

```python
A_roving = 3.62 # [mm**2] 
spacing = 38 # [mm] 
width = B_lower # [mm] 
E_carbon = 200000 # [MPa] 
n_rovings = width / spacing
A_layer = n_rovings * A_roving
n_rovings, A_layer
```

```python
bl1 = ReinfLayer(name='carbon 1', matmod='carbon', z=3, A=A_layer)
bl1.matmod_.trait_set(f_t = 3000, E=E_carbon)
bl2 = ReinfLayer(name='carbon 2', matmod='carbon', z=7, A=A_layer)
bl2.matmod_.trait_set(f_t = 3000, E=E_carbon)
bl3 = ReinfLayer(name='carbon 3', matmod='carbon', z=11, A=A_layer)
bl3.matmod_.trait_set(f_t = 3000, E=E_carbon)
bl4 = ReinfLayer(name='carbon 4', matmod='carbon', z=16+40, A=A_layer)
bl4.matmod_.trait_set(f_t = 3000, E=E_carbon)
```

```python
mc.cross_section_layout.add_layer(bl1)
mc.cross_section_layout.add_layer(bl2)
mc.cross_section_layout.add_layer(bl3)
# mc.cross_section_layout.add_layer(bl4)
```

```python
mc.high_kappa=0.0003
mc.interact()
```

# Self weight

```python
L_m = 3 # m
rho_ton_m3 = 2.400 # t / m3  
rho_kN_m3 = rho_ton_m3 * 10
rho_kN_m3
```

```python
A_c_mm2 = mc.cs_design.cross_section_shape_.get_cs_area()
A_c_m2 = A_c_mm2 / (1000**2)
A_c_m2
```

```python
V_m = A_c_m2 * L_m
m_kN = rho_kN_m3 * V_m * 2 # two interlocked layers
m_kN
```

```python
M_mu_kNm = 7 # kNm
```

$$
 R = \frac{F}{2} \\
 M = R \frac{L}{2} \\
 M = \frac{1}{4} FL \\
 F = 4 \frac{M}{L}
$$

```python
F_mu_kN = 4 * M_mu_kNm / L_m
F_mu_kN
```

```python
F_mu_kN / m_kN
```
